```python
# плотность распределения оценки преподавания от оценки за исследования
plt.scatter(df['teaching'],df['research'],  s = 16, alpha = 0.7)
plt.xlabel('teaching')
plt.ylabel('research')
plt.show()

# зависимость высокая
```

```python
# график рейтинга стран
counts = df['country'].value_counts()
plt.figure(figsize = (12, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', textprops={'fontsize': 6})
plt.title('Countries')
plt.show()
```

```python
# Выборки

# график распределения фичи teaching по генеральной совокупности
df['teaching'].plot(kind = 'hist', label = 'teaching')
```

```python
# Пробуем строить случайную выборку в ~10% от dataframe

random = df.sample(260)
random['teaching'].plot(kind = 'hist', label = 'teaching')
```

```python
# Увеличим процент случайной выборки до 30

random = df.sample(780)
random['teaching'].plot(kind = 'hist', label = 'teaching')
```

```python
# С увеличением процента подгрешность в распределениях заметно снижается
```

```python
# Сделаем выборку по алгоритму

# например каждое 20 значение

alg_sample = df[::20]
alg_sample['teaching'].plot(kind = 'hist', label = 'teaching')

# Видны расхождения
```

```python
# Попробуем уменьшить количество значений

alg_sample = df[::10]
alg_sample['teaching'].plot(kind = 'hist', label = 'teaching')

# почему то выглядит лучше
```

```python
# Выводы:

# Попробовав вероятностоную и невероятностную выборки пришли к выводу что:
# 1. Вероятностная выборка показала себя лучше 
# 2. В невероятностной выборке расхождение росло с количетсвом значений
```
